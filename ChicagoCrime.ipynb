```python
import requests
```

```python
base_api = "https://api.arrayofthings.org/api/"
```

```python
nodes = requests.get(base_api+'nodes?project=chicago').json()['data']
```

```python
nodes
```

```python
#len(nodes)
locations = []
for node in nodes:
    locations.append((node["location"]["geometry"]["coordinates"],node["vsn"]))
```
